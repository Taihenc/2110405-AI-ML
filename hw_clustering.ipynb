````python
# Data clustering

import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from plotly import graph_objects as go
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import plotly.express as px
````

````python
url = 'https://docs.google.com/spreadsheets/d/1TPkXjrsQg_BJXWsMdSUY4Yg0AkZB08De/export?format=csv'
df = pd.read_csv(url)

df
````

````python
df.describe()
````

````python
df.info()
````

````python
# fill na
cleaned_df = df
cleaned_df['time'] = cleaned_df['ช่วงเวลา'].apply(lambda x : float(x.split(':')[0]))
cleaned_df['date'] = cleaned_df['วันที่'].apply(lambda x : float(x.split('/')[0]))

cleaned_df = cleaned_df.fillna(cleaned_df.mean(numeric_only=True))
cleaned_df = cleaned_df.drop(columns=['ช่วงเวลา', 'วันที่', 'No.'])

cleaned_df
````

# Elbow Method for Optimal k

````python
# elbow method
X = cleaned_df.values
wcss = []
for i in range(1,21):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
px.line(x=range(1,21), y=wcss, title='Elbow Method', labels={'x':'Number of clusters', 'y':'WCSS'}).show()
````

````python
# elbow method (scaled)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
wcss_scaled = []

for i in range(1,21):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X_scaled)
    wcss_scaled.append(kmeans.inertia_)
px.line(x=range(1,21), y=wcss_scaled, title='Elbow Method (scaled)', labels={'x':'Number of clusters', 'y':'WCSS'}).show()
````

````python
kmeans = KMeans(n_clusters=6)
kmeans.fit(X_scaled)
````

# Centroid Initialization

````python
# centroids
centroids = kmeans.cluster_centers_
centroids = scaler.inverse_transform(centroids)
centroids_df = pd.DataFrame(centroids, columns=cleaned_df.columns)
centroids_df['cluster'] = centroids_df.index
centroids_df = centroids_df[['cluster'] + list(cleaned_df.columns)]
centroids_df = centroids_df.rename(columns={'cluster':'Cluster'})
centroids_df = centroids_df.round(2)
centroids_df
# plot
fig = go.Figure()
for i in range(centroids_df.shape[0]):
    fig.add_trace(go.Scatter(
        x=centroids_df.columns,
        y=kmeans.cluster_centers_[i],
        mode='lines',
        text=centroids_df.columns,
        textposition='top center',
        name=f'Cluster {i}'
    ))
fig.show()
````

````python
# add cluster to df
cleaned_df['cluster'] = kmeans.labels_

# plot clusters
fig = go.Figure()
for i in range(centroids_df.shape[0]):
    fig.add_trace(go.Scatter(
        x=cleaned_df[cleaned_df['cluster'] == i]['time'],
        y=cleaned_df[cleaned_df['cluster'] == i]['date'],
        mode='markers',
        name=f'Cluster {i}'
    ))
fig.update_layout(
    title='Clusters',
    xaxis_title='Time',
    yaxis_title='Date',
    showlegend=True
)
fig.show()
````

````python
# count for each cluster
cluster_counts = cleaned_df['cluster'].value_counts()
cluster_counts = cluster_counts.reset_index()
cluster_counts.columns = ['Cluster', 'Count']
cluster_counts['Cluster'] = cluster_counts['Cluster'].astype(str)
cluster_counts = cluster_counts.sort_values(by='Cluster')
cluster_counts = cluster_counts.round(2)
cluster_counts
````
